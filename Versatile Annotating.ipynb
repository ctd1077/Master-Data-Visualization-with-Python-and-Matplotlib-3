```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Add an arrow
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
# The xy cordinates shows where the tip of the arrow will point #xy=(np.pi/2, 1)
# The xytext cordinates shows where the arrow will start # xytext=(5,0)
plt.annotate('', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(facecolor='k'))
```

```python
# Add text
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
# The added text is the first argument for annotate function
plt.annotate('1st peak', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(facecolor='k'))
```

```python
# Customize text
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.annotate('Test', xy=(np.pi/2, 1), xytext=(5,0), color='r', weight='bold', style='italic', arrowprops=dict(facecolor='k'))
```

```python
# Customize the arrow (width, headwidth, headlenth, shrink)
# Customize the arrow with arrowprops=dict
# I belive this to be a dict of the arrows properties
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.annotate('Arrow Style', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(facecolor='k', headwidth=50, headlength=40, shrink=0.5))
```

```python
# Customize the arrow (simple arrowstyles)
# Try using 'simple', 'fancy', 'wedge', '|-|', '->', ']-[', '-'
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.annotate('Arrow Style 2', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(arrowstyle='simple'))
```

```python
# Customize the arrow (arrowstyles with attributes)
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.annotate('Arrow Style 3', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(arrowstyle='fancy,head_length=2, head_width=2'))
```

```python
# Customize the arrow (connectionstyle w/ angle3, arc3
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
# plt.annotate('Arrow Style 2', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(arrowstyle='fancy',
                # connectionstyle='arc3, rad=0.5'))
plt.annotate('Arrow Style 4', xy=(np.pi/2, 1), xytext=(5,0), arrowprops=dict(arrowstyle='fancy',
                connectionstyle='angle3, angleA=-20,angleB=30'))
```
