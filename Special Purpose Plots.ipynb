```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

## Polar Axes

```python
nums = np.arange(0,10,0.1)
plt.subplot(111, projection='polar')
plt.plot(nums, nums/10)
plt.plot(nums, np.sin(nums))
plt.plot(nums, np.cos(nums))
```

```python
# Log Axes
# plt.semilogy(nums, np.power(10,10*nums)) # To get log plot for one axis semilogy or semilogx
plt.loglog(nums, np.power(10,10*nums)) # To get log from both axis
# plt.gca(.set_yscale('log')) # you can change the scale after by using the set_yscale or set_xscale functions
plt.gca().grid(True)
```

```python
# Symlog axrs
plt.plot(nums, np.power(10, nums)-100)
# plt.gca().set_yscale('log')
plt.gca().set_yscale('symlog') # Splits into two log plots for positive and negative values
plt.gca().grid(True)
```

```python
# Logistic axes
plt.plot(nums, nums/10)
# plt.gca().set_yscale('log')
plt.gca().set_yscale('logit')
plt.gca().grid(True)
plt.gca().yaxis.set_minor_formatter(mpl.ticker.NullFormatter()) # cleans up 7 tick formatting
```
