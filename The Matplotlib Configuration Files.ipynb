## The Matplotlib Configuration Files

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("bank.csv", sep=';')
```

```python
data.head()
```

```python
plt.plot(data.balance)
plt.xlabel('Customer')
plt.ylabel('Balance')
plt.title('Customer Balance')
plt.legend()
plt.show()
```

```python
mpl.rcParams['lines.linewidth'] = 5
mpl.rcParams['axes.labelcolor'] = 'r'
```
