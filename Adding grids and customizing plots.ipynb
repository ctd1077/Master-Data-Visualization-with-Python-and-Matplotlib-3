```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Data
# Bank and Marketing Data from UCI
# import dataset
data = pd.read_csv('bank.csv', sep=';')
```

```python
data.head()
```

```python
data.balance.plot.hist()
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.show()
```

```python
data.balance.plot.hist(grid=True)
plt.xlabel('Customer')
plt.ylabel('Balance')
plt.title('Customer Balance Distribution');
```

```python
bins=15
rwidth=0.9
data.balance.plot.hist(grid=True, bins=bins, rwidth=rwidth,
                      color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1);
```

```python
# No ticks
bins=15
rwidth=0.9
data.balance.plot.hist(grid=True, bins=bins, rwidth=rwidth,
                      color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1)
plt.xticks([]);
```

```python
bins=15
rwidth=0.9
data.balance.plot.hist(grid=True, bins=bins, rwidth=rwidth,
                      color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1)
plt.xticks(np.arange(0, 1000000, step=100000));
```
