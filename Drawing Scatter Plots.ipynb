```python
# Set Matplotlib to draw plots inline
%matplotlib inline

# Import matplotlib pyplot
import matplotlib.pyplot as plt

# Import numpy
import numpy as np

# Import pandas
import pandas as pd
```

```python
# Online Retail Data Set
# *from https://archive.ics.uci.edu/ml/datasets/online+retail

# Import dataset (excel file
data = pd.read_excel('online-retail.xlsx')
```

```python
# Explore dataset
data.head()
```

```python
# Describe current dataset
data.describe()
```

```python
# Plotting a scatter plot

plt.figure(figsize=(10,6))
plt.scatter(data['UnitPrice'], data['Quantity'])
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.show()
```

```python
# Plotting another scatter plot with different column data

plt.figure(figsize=(10,6))
plt.scatter(data['CustomerID'], data['Quantity'])
plt.xlabel('Customer ID')
plt.ylabel('Quantity')
plt.show()
```

```python
# Plotting another scatter plot with different column data
# And adding in the size of the plot by UnitPrice
plt.figure(figsize=(10,6))
plt.scatter(data['CustomerID'], data['Quantity'], s=data['UnitPrice'])
plt.xlabel('Customer ID')
plt.ylabel('Quantity')
plt.show()
```
