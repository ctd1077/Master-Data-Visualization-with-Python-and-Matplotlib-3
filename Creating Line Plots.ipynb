```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.dpi'] = 150
```

```python
data = pd.read_excel('online-retail.xlsx')
```

```python
# A Simple line plot

plt.plot([1,2,3,4,5,6])
plt.show()
```

```python
# A line plot on our sales data

plt.plot(data['InvoiceDate'], data['Quantity'])
plt.xlabel('InvoiceDate')
plt.ylabel('Quantity')
plt.show()
```

```python
data.loc[data['StockCode'] == 22827]
```

```python
data1 = data.loc[data['StockCode'] == 22827].sort_values('InvoiceDate')
```

```python
# A line plot on our sales data after sorting by InvoiceDate

plt.plot(data1['InvoiceDate'], data1['Quantity'])
plt.xlabel('InvoiceDate')
plt.ylabel('Quantity')
plt.show()
```

```python
data2 = data.loc[data['StockCode'] == 22828].sort_values('InvoiceDate')

plt.plot(data2['InvoiceDate'], data2['Quantity'])
plt.xlabel('InvoiceDate')
plt.ylabel('Quantity')
plt.show()
```

```python
# Sort by InvoiceDate on the whole dataset

data2 = data.sort_values(by=['InvoiceDate'])

plt.plot(data2['InvoiceDate'], data2['Quantity'])
plt.xlabel('InvoiceDate')
plt.ylabel('Quantity')
plt.show()
```
