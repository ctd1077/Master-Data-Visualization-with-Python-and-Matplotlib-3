# Section 3: Building custom plots

```python
# Set matplotlib to draw plots inline
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data
# Bank and Marketing Data from UCI
# *from https://archive.ics.uci.edu/ml/datasets/bank+marketing

#import dataset
data = pd.read_csv('bank.csv', sep=';')
```

```python
# A look at the dataset
data.head()
```

```python
# Add scatter plot using age and balance columns from the dataset
plt.scatter(data['age'], data['balance'])
plt.show()
```

```python
# Add X/Y labels
plt.scatter(data['age'], data['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
# Add Title
plt.scatter(data['age'], data['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.title('Balance by Age')
plt.show()
```

## 3.2: Bulding custom plots using figures

```python
fig = plt.figure()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12,6])
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12,6], facecolor='y')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12,6], facecolor='y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```
