```python
# Pandas Built-in Plotting
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
import pandas as pd
```

# Plotting Series

```python
# Basic Series Plot
p = pd.Series(np.random.randn(1000))
p.plot()
```

```python
# Histogram & KDE
# p.plot.hist(edgecolor='k')
p.plot.kde()
```

```python
# Plotting with Date indices
p = pd.Series(np.random.randn(1000), index=pd.date_range('01/01/1900',periods=1000))
p.plot();
```

```python
# Plotting DataFrames
# cumsum() is important in order to get the line chart
df = pd.DataFrame(np.random.randn(1000,3), columns=['one', 'two', 'three']).cumsum()
df.head()
```

```python
# Basic DataFrame Plot
df.plot()
```

```python
# Plotting data choice
df.plot(x='one', y='two')
```

```python
# Bar Plots
df[0:4].plot.bar()
```

```python
# Tables
df.plot(table=df[0:4])
plt.gca().xaxis.set_visible(False)
```
