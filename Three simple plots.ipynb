```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# basic line plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals))
```

```python
# Axis Labels
plt.plot(xvals, np.sin(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# Multiple Plots
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
```

```python
# Legend & Title
plt.plot(xvals, np.sin(xvals), label=r'$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), label=r'$y = \cos(x)$')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
```

```python
# Line color
plt.plot(xvals, np.sin(xvals), label=r'$y = \sin(x)$', color='c')
plt.plot(xvals, np.cos(xvals), label=r'$y = \cos(x)$', color='m')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
```

```python
# Line style
plt.plot(xvals, np.sin(xvals), label=r'$y = \sin(x)$', color='c', linestyle='--')
plt.plot(xvals, np.cos(xvals), label=r'$y = \cos(x)$', marker='o', linestyle='none')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
```

```python
# Short format code
plt.plot(xvals, np.sin(xvals), 'c+--', label=r'$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), label=r'$y = \cos(x)$')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
```

```python
# Axis Limits
plt.plot(xvals, np.sin(xvals), 'c--', label=r'$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), 'mo', label=r'$y = \cos(x)$')
plt.ylim(0,1)
plt.xlim(0,1)
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
```

```python
# Basic Histogram plots
```

```python
# Generate 100 random numbers
rands = np.random.normal(size=1000)
```

```python
# Histogram
plt.hist(rands)
plt.xlabel(r'$x$')
plt.ylabel(r'$Pr(x)$')
```

```python
# Generate a smoothed, gaussian random field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
print(rands2d.shape)
```

```python
# Show Iamge
plt.imshow(rands2d, cmap='inferno')
plt.colorbar()
```
