```python
# Controlling axes and ticks
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

## Axis Objects

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.yaxis
```

```python
# Grid Lines
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.yaxis.grid(True)
```

```python
# Location of Axis
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.yaxis.set_ticks_position('right')
```

## Ticks

```python
# Get tick locations
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.yaxis.get_ticklocs()
```

```python
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_major_locator(ticker.MaxNLocator(3))
```

```python
# Prune Upper/Lower ticks
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_major_locator(ticker.MaxNLocator(prune='lower'))
# you can remove your 'lower' most value, 'upper' most, or 'both'
```

```python
# Integer only ticks
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
```

```python
# Manually selected ticks with FixedLocator
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_major_locator(ticker.FixedLocator([0.1,0.7,0.9]))
# you can also pass FixedLocator and empty array and get 0 ticks
```

```python
# Minor ticks
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_minor_locator(ticker.MaxNLocator(20))
```

## Labels

```python
# Label position
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_ylabel('y label')
ax.yaxis.set_label_position('right')
# set_label_position('left') is the default label position
```

```python
# Label coordinates
from matplotlib import ticker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_ylabel('y label')
ax.yaxis.label.set_y(0)
```
