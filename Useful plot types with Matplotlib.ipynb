```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

## Fill & Fill Between

```python
# Fill
xvals = np.arange(0,10,0.1)
plt.fill(xvals, np.sin(xvals))
```

```python
# Fill color
plt.fill(xvals, np.sin(xvals), color='c')
plt.fill(xvals, np.sin(xvals), hatch='x')
```

```python
# Fill between
# Takes two each non key word arguements
# Two lines, one grows to the right and (xvals+np.sin(xvals))
# the other one shrinks to the right (np.sin(xvals)-xvals)
plt.fill_between(xvals, xvals+np.sin(xvals), np.sin(xvals)-xvals, edgecolor='k')
```

## Hexbin & Hist2D

```python
randx = np.random.normal(size=int(1e5))
randy = np.random.normal(size=int(1e5))
```

```python
plt.plot(randx, randy, 'k')
```

```python
# Hexbin
plt.hexbin(randx, randy)
```

```python
# Bins
plt.hexbin(randx, randy, bins=5)
```

```python
# Mincnt
plt.hexbin(randx, randy, mincnt=1)
```

```python
# Hist2d
# In general it is better to use hexbin not hist2d
plt.hist2d(randx, randy);
```
