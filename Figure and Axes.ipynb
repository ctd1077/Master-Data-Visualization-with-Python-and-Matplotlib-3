```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Generate a figure & axes
# fig.add_axes([left, bottom, width, height])
# fig = plt.figure(dpi=10)
# fig = plt.figure(figsize[1,1]) # In inches [1,1] 1 by 1 inch figure
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
fig.set_size_inches([1,1]) # after the figure is created you can change the sizes this way.
```

```python
# Frames with linewidth and edgecolor
# Think of the figure as a container
fig = plt.figure(linewidth=1, edgecolor='b')
ax = fig.add_axes([0,0,1,1])
```

```python
# Mutliple axes
# And enbedding a figure inside another figure # Useful when you have a subset of data
fig = plt.figure()
# ax = fig.add_axes([0,0,0.1,0.1])
ax = fig.add_axes([0,0,1,1])
ax = fig.add_axes([0,0.5,0.1,0.1])
```

```python
# Tight Layout
plt.subplots(2,2);
plt.tight_layout()
```

```python
# Clearing figures & Axes
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,0.5])
ax1.plot(range(0,10), range(0,10))
ax2 = fig.add_axes([0,0.5,1,0.5])
ax2.plot(range(0,10), range(0,10))
#fig.clf() # Clears the whole figure
# ax1.cla() # Clears the individual axis
ax2.cla() # Or clear both ans still see the figure on the screen
```

```python
# Get current Axes & Figure
plt.plot()
print(plt.gcf()) # Get current figure
print(plt.gca()) # Get current axes
```
