```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.dpi'] = 150
```

```python
data = pd.read_csv('auto-mpg.csv', index_col='name')
```

```python
data.head()
```

```python
# A Simple Histogram

plt.hist(data.mpg)
plt.title('MPG')
plt.show()
```

```python
# Customizations

data.mpg.plot.hist(grid=True, bins=15, rwidth=0.9,
                  color='#607c8e')
plt.title('Miles per Gallon')
plt.xlabel('MPG')
plt.ylabel('NO. of vehicles')
```

```python
# Comparisons
fig, ax = plt.subplots()
data[['mpg', 'horsepower']].plot.kde(ax=ax, legend=False, title='Histogram A vs B')
data[['mpg', 'horsepower']].plot.hist(density=True, ax=ax)
ax.set_ylabel('Probability')
ax.grid(axis='y')
```
