```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
%matplotlib notebook
```

```python
from matplotlib.animation import FuncAnimation
```

```python
# Basic FuncAnimation
# How is the argument chosen?
nums = np.arange(0,10,0.1)
sin, = plt.plot(nums, np.sin(nums))
def update(i):
    sin.set_ydata(np.sin(nums+i))
    plt.title(i) # To see the values that are being passed
FuncAnimation(plt.gcf(), update) # First arguement is gcf second is the function we created
```

```python
# Specifying our own values for the argument with frames
nums = np.arange(0,10,0.1)
sin, = plt.plot(nums, np.sin(nums))
frames=np.arange(0,2*np.pi,0.1)
def update(i):
    sin.set_ydata(np.sin(nums+i))
    plt.title(i)
FuncAnimation(plt.gcf(), update, frames=frames)
```

```python
# Interval between frames
nums = np.arange(0,10,0.1)
sin, = plt.plot(nums, np.sin(nums))
frames = np.arange(0,2*np.pi,0.1)
def update(i):
    sin.set_ydata(np.sin(nums+i))
    plt.title(i)
FuncAnimation(plt.gcf(), update, frames=frames, interval=60)
```

```python
# Repeat, pause after final frame
nums = np.arange(0,10,0.1)
sin, = plt.plot(nums, np.sin(nums))
frames=np.arange(0,2*np.pi,0.1)
def update(i):
    sin.set_ydata(np.sin(nums+i))
    plt.title(i)
FuncAnimation(plt.gcf(), update, frames=frames, interval=60, repeat_delay=2e3)
```

```python
# Saving the animation as a mp4 video
nums = np.arange(0,10,0.1)
sin, = plt.plot(nums, np.sin(nums))
frames=np.arange(0,2*np.pi,0.1)
# Set up formatting for the movie files
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
def update(i):
    sin.set_ydata(np.sin(nums+i))
    plt.title(i)
anim = FuncAnimation(plt.gcf(), update, frames=frames, interval=50)
anim.save('sine.mp4', writer=writer)
```
