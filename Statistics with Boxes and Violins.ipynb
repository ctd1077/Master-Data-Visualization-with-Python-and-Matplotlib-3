```python
# Statistics with Boxes and Violins
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

## Box Plots

```python
# Generate some Normal distributions with different properties
rands1 = np.random.normal(size=500)
rands2 = np.random.normal(scale=2, size=500)
rands3 = np.random.normal(loc=1, scale=0.5, size=500)
gaussians = (rands1, rands2, rands3)
```

```python
# Basic Boxplot
plt.boxplot(gaussians);
```

```python
# Labels
plt.boxplot(gaussians, labels=('First', 'Second', 'Third'));
```

```python
# Box widths
# You can pass one width value for all items or pass a tuple for different widths
plt.boxplot(gaussians, widths=(0.1, 0.5, 0.7));
```

```python
# Horizontal boxes w/ vert
plt.boxplot(gaussians, vert=False);
```

```python
# Outliners w/ sym
plt.boxplot(gaussians, sym='o');
```

## Violin Plots

```python
# Some other kinds of distrubutions
rands4 = np.random.lognormal(size=500, sigma=0.5)
rands5 = np.random.pareto(size=500, a=5)
dists = (rands1, rands4, rands5)
```

```python
# Basic Violin plot
plt.violinplot(dists);
```

```python
# Show Median & Means
plt.violinplot(dists, showmeans=True, showmedians=True);
```

```python
# Getting the components
vplot = plt.violinplot(dists, showmeans=True, showmedians=True)
# print(vplot.keys()) 
vplot['cmedians'].set_color(('r', 'c', 'w')) # You can pass a tuple of colors
vplot['bodies'][0].set_color('k') # bodies is actually a list
vplot['bodies'][1].set_color('g')
```
