```python
# Differentiating Line and Scatter Plots
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basic line plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# different line styles
idx = 0
for marker in ('-', '--', ':', '-.'):
    plt.plot(xvals, np.sin(xvals)+idx,  ''.join(('k',marker)))
    idx += 1
```

```python
# Markers
plt.plot(xvals, np.sin(xvals), 'k.--')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Marker frequency (markevery)
xvals = np.arange(0,10,0.01)
plt.plot(xvals, np.sin(xvals), 'k.', markevery=10)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Marker size (markersize or ms)
plt.plot(xvals, np.sin(xvals), 'r.', ms=40)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Different Markers in matplotlib
idx = 0
for marker in ('.', ',', 'o', 'v', '^', '>', '<','1', '2', '3', '4', 's', 'p',
               '*', 'h', 'H','+', 'x', 'D', 'd', '|', '_'):
    plt.plot(idx, 0, ''.join(('r', marker)))
    idx += 1
plt.xlim(-0.5, idx+0.5)
```

```python
# Latex Marker
xvals = np.arange(0,10,0.01)
plt.plot(xvals, np.sin(xvals), color='c', linestyle='none', marker=r'$\Sigma$', markevery=15)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Line Width
plt.plot(xvals, np.sin(xvals), linewidth=.1)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# zorder (Which line is closest to the foreground)
plt.plot(xvals, np.sin(xvals), lw=8, zorder=1)
plt.plot(xvals+1, np.sin(xvals), lw=8, zorder=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Transparency
plt.plot(xvals, np.sin(xvals), lw=8, alpha=0.5)
plt.plot(xvals+1, np.sin(xvals), lw=8, alpha=0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# array order matters!
# This works when plotting indivual markers ('k.')
shuffled_xvals = np.random.permutation(xvals)
plt.plot(shuffled_xvals, np.sin(shuffled_xvals), 'k.-')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

## scatter()

```python
xvals = np.arange(0,10,0.1)
plt.scatter(xvals, np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Colours
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals))
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals), cmap='inferno')
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Set Edge Color
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals), edgecolor='k')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Size
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals), s=np.power(xvals, 3), edgecolor='k')
plt.colorbar()
plt.xlim(-2,15)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

## Heads up about speed!
* When you have large datasets, plot method can be a lot faster to use.

```python
%%timeit -n3 -r3 plt.plot(np.random.uniform(size=int(1e6)), np.random.uniform(size=int(1e6)), 'k')
plt.clf()
```

```python
%%timeit -n3 -r3 plt.scatter(np.random.uniform(size=int(1e6)), np.random.uniform(size=int(1e6)), c='k', marker='.')
plt.clf()
```
