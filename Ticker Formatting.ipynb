```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker 
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

## Modifying tick labels & formatting manually

```python
# Manuallhy set tick labels
index = ('one', 'two', 'three', 'four')
plt.plot()
plt.ylim(1,4)
plt.yticks([1,2,3,4], index)
```

```python
# Rotating ticks
index = ('one', 'two', 'three', 'four')
plt.plot()
plt.ylim(1,4)
plt.yticks(rotation=45)
# any degree of rotation will work, try (-90)
```

```python
# Tick Color
index = ('one', 'two', 'three', 'four')
plt.plot()
plt.ylim(1,4)
plt.yticks(color='r')
```

## Using formatters

```python
# NullFormatter (no labels)
plt.plot()
plt.gca().yaxis.set_major_formatter(ticker.NullFormatter())
```

```python
# Log vs Linear formatting
plt.loglog()
plt.ylim(1,1e7)
#plt.gca().yaxis.set_major_formatter(ticker.ScalarFormatter())
#plt.gca().yaxis.set_major_formatter(ticker.LogFormatter())
# Default option what you see is ticker.LogFormatterMathtext()
```

```python
# EngFormatter W/Unit
plt.loglog()
plt.ylim(1,10000)
plt.gca().yaxis.set_major_formatter(ticker.EngFormatter(unit='Hz'))
# This can be very useful
```

```python
# FormatDtrFormatter
plt.plot()
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('This tick value %3.2f'))
# %3.2f this is saying, 'give me three digits with 2 after the decimal point'
```

```python
# FuncFormatter
func = lambda x,pos:r'$\sqrt{\rm tick}$ is %3.2f' % np.sqrt(x)
# x is the value of the tick and pos is the position of the tick
plt.plot()
plt.ylim(1,10)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(func))
```
