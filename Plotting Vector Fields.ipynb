```python
# Plotting Vector Fields
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
plt.rcParams['image.cmap']='hsv'
```

```python
# Generate a vector field with a gradient
from scipy.ndimage.filters import gaussian_filter
x = np.arange(0,10,0.5)
y = np.arange(0,10,0.5)
phi = gaussian_filter(np.random.uniform(size=(20,20)), sigma=5)
plt.subplot(141)
plt.imshow(phi, interpolation='none')
plt.title('$\Phi$')
plt.subplot(142)
plt.imshow(np.gradient(phi)[0], interpolation='none')
plt.title('$\partial_x\Phi$')
plt.subplot(143)
plt.imshow(np.gradient(phi)[1], interpolation='none')
plt.title('$\partial_y\Phi$')
plt.subplot(144)
plt.title(r'$\|\nabla \Phi\|$')
plt.imshow(np.linalg.norm(np.gradient(phi), axis=0), interpolation='none')
plt.gcf().set_size_inches(8,4)
```

```python
# Vectors with quiver
plt.imshow(phi, extent=(0,10,0,10), interpolation='none', origin='lower')
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0])
```

```python
# Vector scale
# plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], scale=1)
# You get bigger vectors with small values of scale
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], scale=0.1)
```

```python
# Units (x,y,height,width(default),inches)
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], scale_units='inches', scale=0.1)
```

```python
# Vector width
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], width=0.001)
```

```python
# Vector color
# plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color='r')
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], phi)
```

```python
# Pivot (where the vector goes)
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color='r', pivot='tail')
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color='k', pivot='mid')
plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color='b', pivot='tip')
```

```python
# Key
q = plt.quiver(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color='c')
plt.quiverkey(q, 0.1, 0.9, 0.01, 'Gradient = 0.01', color='b')
```

```python
# Stream plot
plt.streamplot(x,y, np.gradient(phi)[1], np.gradient(phi)[0])
```

```python
# Stream density
plt.streamplot(x,y, np.gradient(phi)[1], np.gradient(phi)[0], density=2)
```

```python
# Stream plot linewidth
delta = phi-np.mean(phi)
# plt.streamplot(x,y, np.gradient(phi)[1], np.gradient(phi)[0], color=phi)
plt.streamplot(x,y, np.gradient(phi)[1], np.gradient(phi)[0], linewidth=delta)
```
