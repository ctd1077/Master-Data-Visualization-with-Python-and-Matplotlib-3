## Bar Chart

```python
# Load numpy for math/array opperations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# basic bar plot
nums = np.random.uniform(size=10)
plt.bar(np.arange(10), height=nums)
```

```python
# horizontal bar plot
plt.barh(np.arange(10), width=nums)
```

```python
# Bar Color
plt.bar(np.arange(10), nums, color='limegreen', edgecolor='k')
```

```python
# Bar alignment
plt.bar(np.arange(10), nums, align='center')
```

```python
# Hatch & fill
plt.bar(np.arange(10), nums, hatch='x', color='none', edgecolor='k')
```

```python
# Bar Width ()
plt.bar(np.arange(10), nums, color='limegreen', edgecolor='k', width=1)
```
