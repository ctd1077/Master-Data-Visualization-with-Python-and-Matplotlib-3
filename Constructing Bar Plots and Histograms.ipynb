## Bar Chart

```python
# Load numpy for math/array opperations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# basic bar plot
nums = np.random.uniform(size=10)
plt.bar(np.arange(10), height=nums)
```

```python
# horizontal bar plot
plt.barh(np.arange(10), width=nums)
```

```python
# Bar Color
plt.bar(np.arange(10), nums, color='limegreen', edgecolor='k')
```

```python
# Bar alignment
plt.bar(np.arange(10), nums, align='center')
```

```python
# Hatch & fill
plt.bar(np.arange(10), nums, hatch='x', color='none', edgecolor='k')
```

```python
# Bar Width (.8 si what you see above, 1 is touching, 1+ is overlaping)
plt.bar(np.arange(10), nums, color='limegreen', edgecolor='k', width=1)
```

```python
# Edge Color
plt.bar(np.arange(10), nums, color='limegreen', edgecolor='none', width=1)
```

```python
# Bottom alignment ((ararry value)bottom=0.1*nums) = this method will show you error bars around certain values
plt.bar(np.arange(10), nums, bottom=0.1*nums)
```

## hist() - The most useful chart or bar chart for data scientist

```python
# Basic histogram
rands = np.random.normal(size=int(1e6))
plt.hist(rands) # if you don't want to show the text/ararrys just use a (;) example: plt.hist(rands);
```

```python
# Return type
print(plt.hist(rands))
plt.clf()
```

```python
# cla()   # Clear axis
# clf()   # Clear figure
# close() # Close a figure window
```

```python
# Range
plt.hist(rands, range=(0,1), edgecolor='k');
```

```python
# Bin Choice
plt.hist(rands, bins=[0,1,2,3], edgecolor='k')
```

```python
# HistType
plt.hist(rands, bins='auto')
```

```python
# Normalization
plt.hist(rands, bins='auto', edgecolor='k', histtype='stepfilled');
```

```python
# Cumulative Histogram
#plt.hist(rands, bins='auto', cumulative=True);
plt.hist(rands, bins='auto', cumulative=-1, normed=True);
```

```python
# Multiple Histograms
# Pass hist a ararry or tuple of values to be plotted (using the stacked keyword)
plt.hist((rands, 0.5*rands), bins='auto', histtype='stepfilled', stacked=True);
```
