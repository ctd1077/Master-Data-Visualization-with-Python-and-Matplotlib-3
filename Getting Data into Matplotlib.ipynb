```python
# Set matplotlib to draw plots inline
%matplotlib inline

#import matplotlib pyplot
import matplotlib.pyplot as plt

# import numpy
import numpy as np

# import pandas
import pandas as pd
```

```python
# Passing 1st data to plot
x = [0,1,2,3,4,5]
y = [0,10,20,30,40,50]

plt.plot(x,y)
plt.show
```

```python
# Using numpy to process data and then pass it to matplotlib for plotting
x = np.arange(0,100)
y = 9 * x +7

plt.plot(x,y)
plt.show()
```

```python
# Gross World Product Data since 1970 - from wikipedia
data = pd.read_csv('gwp.csv')
```

```python
# Explore data
data
```

```python
# Plotting pandas dataframe by passing dataframes columns as x and y axis
plt.plot(data['Year'], data['GWP'])
plt.show()
```

```python
# Data Visualization by calling plot function directly on the dataframe

# Set Year as index
data1 = data.set_index('Year')
```

```python
data1
```

```python
# Creating a plot by calling plot directly on the dataframe

data1['GWP'].plot() # The column called is the y axis then the x axis is set to the index
plt.show()
```
