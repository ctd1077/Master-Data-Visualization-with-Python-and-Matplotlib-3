```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

## Subplot

```python
# Generate a ndarray of nums from 0 to 10 with increments of 0.1
# arange(Start, Stop, Step)
nums = np.arange(0,10,0.1)

# Subplot (Rows, Columns, (plot in columns or rows))
# plt.subplot(1,2,1)
plt.subplot(2,1,1)
plt.plot(nums, np.sin(nums))
# plt.subplot(1,2,2)
plt.subplot(2,1,2)
plt.plot(nums, np.cos(nums))
# subplot could also be written as plt.subplot(211) or plt.subplot(212)
```

```python
# Subplot labels (the 'current axis' state)
plt.subplot(2,1,1)
plt.plot(nums, np.sin(nums))
plt.ylabel('sin')
plt.subplot(2,1,2)
plt.plot(nums, np.cos(nums))
plt.ylabel('cos')
plt.xlabel('X');
```

## Subplots

```python
fig, axes = plt.subplots(2,2)
print(axes.shape)
plt.clf()
```

```python
# Looping with the axis array
fig, axes = plt.subplots(2,2)
i = 1
for ax in axes.flatten(): # We use flatten() to turn axes into a unidimensional array
    ax.plot(nums, np.power(np.sin(nums), i))
    i += 1
```
