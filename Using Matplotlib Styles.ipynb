```python
# Using Matplotlib Styles
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
data = pd.read_csv('bank.csv', sep=';')
```

# Matplotlib Styles

```python
plt.style.available
```

```python
# Simple plot
plt.plot(data.balance)
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance')
plt.show()
```

```python
plt.style.use('ggplot')

plt.plot(data.balance)
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance')
plt.show()
```

```python
plt.style.use('seaborn')

plt.plot(data.balance)
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance')
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.plot(data.balance)
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance')
plt.show()
```
