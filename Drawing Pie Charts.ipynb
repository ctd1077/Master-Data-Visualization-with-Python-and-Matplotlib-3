```python
# Drawing Pie Charts
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.dpi'] = 150
```

```python
data = pd.read_csv('auto-mpg.csv', index_col='name')
```

```python
# Data
data.head()
```

```python
# Group by the data we want to show in the pie chart
cyl = data.groupby('cylinders').size()
cyl
```

```python
# Prep the data to plt by seperating the keys and values
```

```python
cylinders = cyl.keys().tolist()
cylinders
```

```python
cars = cyl.values.tolist()
cars
```

```python
# Plot the keys and values
# The cars list is the number of cars for the label cylinders
plt.pie(cars, labels=cylinders, startangle=90, autopct='%.1f%%')
plt.title('Number of cars by Cylinders')
plt.show()
```

```python
fig1, ax1 = plt.subplots()

ax1.pie(cars, labels=cylinders, autopct='%1.1f%%', startangle=90)

# Draw Circle
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is draw as a circle
ax1.axis('equal')
plt.tight_layout()
plt.show()
```
