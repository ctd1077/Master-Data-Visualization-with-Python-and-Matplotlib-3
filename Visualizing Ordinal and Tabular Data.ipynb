```python
# Visualizing Ordinal and Tabular Data
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

## Pie Charts

```python
# Basic Pie Chart
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels);
plt.gca().set_aspect('equal')
```

```python
# Explode!
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
explode = (0, 0.2, 0)
plt.pie(fracs, explode=explode, labels=labels);
plt.gca().set_aspect('equal')
```

```python
# Colors
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, colors=('b', 'c', 'r'), labels=labels);
plt.gca().set_aspect('equal')
```

```python
# Appearance tweaks (shadow, radius)
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels, shadow=True);
# plt.pie(fracs, labels=labels, radius=0.5);
plt.gca().set_aspect('equal')
```

```python
# Show percentages w/ autopct
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels, autopct='%2.0f %%') 
# The %2(Means two digits) .0(no digits after (.) f(format) and %% (means to add %))
plt.gca().set_aspect('equal')      
```

## Tables

```python
# Add a table
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], colLabels=labels, rowLabels=['Fraction'])
plt.gca().set_aspect('equal')  
```

```python
# Change location w/ loc
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], rowLabels=['Factions'], colLabels=labels, loc='left')
plt.gca().set_aspect('equal')  
```

```python
# Change colors w/ rowColors, cellColors & colColors
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], rowLabels=['Factions'], colLabels=labels, colColours=['b','g','r'])
# plt.table(cellText=[fracs], rowLabels=['Factions'], colLabels=labels, cellColours=[['b','g','r']])
plt.gca().set_aspect('equal')  
```
