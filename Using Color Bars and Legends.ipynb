```python
# Using Color Bars and Legends
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

## Legends

```python
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tan(nums), label='tan')
plt.legend()
plt.ylim(-2,2)
```

```python
# Location
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tan(nums), label='tan')
plt.legend(loc='upper left')
plt.ylim(-2,2)
```

```python
# Number of columns
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tan(nums), label='tan')
plt.plot(nums, np.sinh(nums), label='sinh')
plt.plot(nums, np.cosh(nums), label='cosh')
plt.plot(nums, np.tanh(nums), label='tanh')
plt.legend(ncol=3, loc='upper right')
plt.gcf().set_size_inches(6,2)
plt.ylim(-2,2)
```

```python
# Title
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tan(nums), label='tan')
plt.legend(title='Trig Function', loc='best')
plt.ylim(-2,2)
```

```python
# Appearance (frameon, fancybox, shadow, fontsize)
plt.plot(nums, np.sin(nums), label='sin')
plt.plot(nums, np.cos(nums), label='cos')
plt.plot(nums, np.tan(nums), label='tan')
#plt.legend(frameon=False)
#plt.legend(fancybox=True)
plt.legend(shadow=True)
plt.legend(fontsize=5)
plt.ylim(-2,2)
```

## Colorbars

```python
# Generage a smoothed, gaussian ransom field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
plt.imshow(rands2d)
plt.colorbar()
```

```python
# Orientation
plt.imshow(rands2d)
plt.colorbar(orientation='horizontal')
```

```python
# label
# set varible cbar
plt.imshow(rands2d)
cbar = plt.colorbar(orientation='horizontal')
cbar.set_label('Gaussian Random Field')
```
