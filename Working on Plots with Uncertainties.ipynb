```python
# Working on Plots with Uncertainties
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

## Uncertainties on Line & Scatter Plots

```python
nums = np.arange(0,10,0.5)
xerrors = np.random.uniform(size=20)
yerrors = np.random.uniform(size=20)
```

```python
# Basic errorbar plot
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors)
```

```python
# Errorbar Color
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, ecolor='r')
```

```python
# Errorbar line width
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, elinewidth=.1, ecolor='k')
```

```python
# Errorbar cap size
# Default setting is 0. You wont be able to see them unless
# the arguement is greatter than 0
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, elinewidth=.1, capsize=5)
```

```python
plt.bar(nums, np.sin(nums)**2, width=0.2, xerr=xerrors, yerr=yerrors)
```
