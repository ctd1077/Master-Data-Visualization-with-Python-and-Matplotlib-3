```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Standard Subplot
#subplot(#ofRows, #ofColumns, )
plt.subplot(311)
plt.subplot(312)
plt.subplot(313)
```

```python
# Subplot2grid
# colspan says plt should span the number of columns it equals
# you can do the same thing with row span but the plt should start
# at the 0 position
plt.subplot2grid((2,2),(0,0))#, colspan=2)
plt.subplot2grid((2,2),(1,0))
plt.subplot2grid((2,2),(0,1), rowspan=2)
```

```python
# Explicit GrisSpec
gs = mpl.gridspec.GridSpec(2,3)
plt.subplot(gs[0,:]) # Starts at 0 position to the end of gridspec
plt.subplot(gs[1,1])
```

```python
# Adjusting GridSpec attributes: w/h space
gs = mpl.gridspec.GridSpec(2,2, wspace=0.5)
plt.subplot(gs[0,0])
plt.subplot(gs[0,1])
plt.subplot(gs[1,0])
plt.subplot(gs[1,1])
```

```python
# Adjusting GridSpec attributes: width/height ratios
gs = mpl.gridspec.GridSpec(2,2, width_ratios=(1,1.5)) # 2nd column is 1.5 the size of the first
plt.subplot(gs[0,0])
plt.subplot(gs[0,1])
plt.subplot(gs[1,0])
plt.subplot(gs[1,1])
```
