```python
# Create data
x = np.random.randn(4096)
y = np.random.randn(4096)

# Create hearmap
heatmap, xedges, yedges = np.histogram2d(x, y, bins=(64, 64))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Plot Heatmap
plt.clf()
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(heatmap, extent=extent)
plt.show()
```
