```python
# Basic Box Plot
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.dpi'] = 150
```

```python
data = pd.read_csv('auto-mpg.csv', index_col='name')
```

```python
# Basic boxplot
plt.boxplot(data['mpg'])
plt.show()
```

```python
# Notch plot
plt.figure()
plt.boxplot(data['mpg'], 1)
plt.show()
```

```python
# Change Outlier point symbols
plt.figure()
plt.boxplot(data.mpg, 0, 'rD')
plt.show()
```

```python
# Don't show outlier points
plt.figure()
plt.boxplot(data.mpg, 0, '')
plt.show()
```

```python
# Horizontal boxes
plt.figure()
plt.boxplot(data.mpg, 0, '', 0)
plt.show()
```

```python
# Change whisker length
plt.figure()
plt.boxplot(data.mpg, 0, 'rs', 0, 0.75)
plt.show()
```
