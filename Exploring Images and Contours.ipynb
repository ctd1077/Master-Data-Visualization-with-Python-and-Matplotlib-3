```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Generate a smoothed, gaussian ransom field
from scipy.ndimage.filters import gaussian_filter
randsd = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
# Basic imshow
plt.imshow(randsd)
plt.colorbar()
```

```python
# Extent & LRBT 
# The extent agurement takes four variable for x, y axis
# Going from left, right (x axis) then bottom, top (y axis)
plt.imshow(randsd, extent=[0,500,0,500])
plt.colorbar()
```

```python
# Min & Max
plt.imshow(randsd, vmax=0.05, vmin=-0.05,  cmap='inferno')
plt.colorbar()
```

```python
# Aspect ratio 
# Try aspect=2
# Default value is 1
plt.imshow(randsd, aspect=.2)
```

```python
# interpolation
small = np.arange(0,4,1).reshape(2,2)
#print(small)
#plt.imshow(small)
#plt.imshow(small, interpolation='gaussian')
plt.matshow(small)
```

```python
# Basic Contour
srand2d = gaussian_filter(randsd, sigma=100)
#plt.imshow(srand2d)
plt.contour(srand2d)
plt.colorbar()
```

```python
# Contour levels
plt.contour(srand2d, levels=[0, 0.003])
```

```python
# Number of Contours
plt.contour(srand2d, 30)
```

```python
# Contour Colors
plt.contour(srand2d, 10, colors=['k', 'c'])
```

```python
# Filled Contours
plt.contourf(srand2d, 10)
```

```python
# Hatches
plt.contourf(srand2d, 10, colors='none', hatches=['x', '/'])
```

```python
# Line Styles and widths 
# Didn't notice a change with widths

# plt.contour(srand2d, widths=[1,5])
plt.contour(srand2d, linestyles=[':','--','-'])
```
