```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
def setup_map(m):
    m.drawcoastlines(zorder=-1)
    m.fillcontinents(color='coral',lake_color='aqua', zorder=-1)
    # draw parallels and merdians
    m.drawparallels(np.arange(-90.,120.,30.))
    m.drawmeridians(np.arange(0.,420.,60.))
    m.drawmapboundary(fill_color='aqua')
    plt.gcf().set_size_inches(12,10)
```

```python
# Projecting with and without latlon
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
setup_map(m)
HD = (8.7, 49.5)
m.scatter(*HD, latlon=True, color='k', marker='x') # Not Working correctly
# This line plots the x but the continents disappear
plt.show()
```

```python
# Plotting lines
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
setup_map(m)
m.plot(np.linspace(-10,60,2), 45*np.ones(2), 'g-', latlon=True, lw=3)
plt.show()
```

```python
# Terminator 2: Judgement Datetime (nightshade)
from datetime import datetime
m= Basemap()
setup_map(m)
m.nightshade(datetime.utcnow())
plt.show()
```
